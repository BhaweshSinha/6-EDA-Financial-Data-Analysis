### Adding the new column onto the dataframe named 'Cap_Sales ratio'
CSR = M-cap/Sales_qtr

```python
p1 = [x for x in df['Mar Cap - Crore']]
```

```python
p2 = [y for y in df['Sales Qtr - Crore']]
```

```python
main = []
for i in range(len(p1)):
    main.append(p1[i]/p2[i])
df['Cap_Sales ratio'] = main
```
